```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
yf.Ticker("AAPL")
```

```python
yf.Ticker("AAPL").info
```

```python
tickers=["AAPL","MSFT","IBM"]
infos=[]
for i in tickers:
    infos.append(yf.Ticker(i).info)
    infos
```

```python
infos
```

```python
fundamentals=["dividendyield","marketcap","beta","forwardPE"]
df=pd.DataFrame(infos)
df
```

```python
df=df.set_index("symbol")
df
```

```python
df.columns.isin(fundamentals)
```

```python
df.columns[df.columns.isin(fundamentals)]
```

```python
df[df.columns[df.columns.isin(fundamentals)]]
```

```python
plt.bar(df.index,df.dividendYield,color=("k","orange","cyan"))
plt.show
```
